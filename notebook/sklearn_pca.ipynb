```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
from sklearn import decomposition
from sklearn import datasets

np.random.seed(5)

centers = [[1,1],[-1,-1],[1,-1]]
iris = datasets.load_iris()
X = iris.data
y = iris.target

fig = plt.figure(1, figsize=(4,3))
plt.clf()
ax = Axes3D(fig, rect=[0,0,.95,1], elev=48, azim=134)

plt.cla()
pca = decomposition.PCA(n_components=3)
pca.fit(X)
X_t = pca.transform(X)
```

```python
print(pca.explained_variance_ratio_)
print(pca.explained_variance_)
```

```python
print(pca.n_components_)
y_c = np.choose(y, [1, 2, 0]).astype(np.float)
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y_c, cmap=plt.cm.nipy_spectral,
           edgecolor='k')
ax.w_xaxis.set_ticklabels([])
ax.w_yaxis.set_ticklabels([])
ax.w_zaxis.set_ticklabels([])

plt.show()
```

```python
fig = plt.figure(1, figsize=(4,3))
plt.clf()
ax = Axes3D(fig, rect=[0,0,.95,1], elev=48, azim=134)

plt.cla()
y_c = np.choose(y, [1, 2, 0]).astype(np.float)
ax.scatter(X_t[:, 0], X_t[:, 1], X_t[:, 2], c=y_c, cmap=plt.cm.nipy_spectral,
           edgecolor='k')
ax.w_xaxis.set_ticklabels(['x'])
ax.w_yaxis.set_ticklabels(['y'])
ax.w_zaxis.set_ticklabels(['z'])

plt.show()
```
